## How many apps are there with rating 5 ?

```python
c=0
for i in data['Rating']:
    if (i== 5.0):
        c += 1
        
print("There are",c,"Applications with rating 5")        
```

## Q3. How many apps are there with rating between 4 - 4.5?

```python
c = 0

for i in data['Rating']:
    if (i >= 4.0 and i <= 4.5):
        c += 1
        
print("There are",c,'apps with rating between 4 - 4.5')
```

## Q4. Average App Reviews


```python
s=0
for i in data['Reviews']:
    s += int(i)

print("Average of App Reviews is:",int(s/len(data['Reviews'])))
```
